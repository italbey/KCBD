1.Гістограма віку (age):
Побудуйте гістограму віку клієнтів, використовуючи Seaborn.

```python
sns.histplot(data=df, x='Age', bins=15, kde=True)
plt.title('customers age')
plt.xlabel('age')
plt.ylabel('Number of customers')
plt.show()
```

2.Порівняння розподілу покупок за статтею (item_purchased):
Створіть стовпчасту діаграму або графік розподілу, щоб порівняти, які
товари покупають клієнти найчастіше.

```python
items = df['Item Purchased'].value_counts().reset_index()
items.columns = ['Item', 'Number of purchases']
plt.figure(figsize=(12,6))
sns.barplot(data=items.head(10), x='Number of purchases', y='Item', palette='viridis')
plt.title('popular items')
plt.xlabel('Number of purchases')
plt.ylabel('Items')
plt.show()
```

3.Залежність між віком та сумою покупок (purchase_amount):
Використовуйте scatter plot, щоб візуалізувати залежність між віком та
сумою покупок.

```python
plt.figure(figsize=(12,8))
sns.scatterplot(data=df, x='Age', y='Purchase Amount (USD)', alpha=0.6, color='blue')
plt.title('Correlation between age and purchase amount')
plt.xlabel('Age')
plt.ylabel('Purchase amount')
plt.show()
```

7.Сезонність покупок (season):
Побудуйте стовпчату діаграму, яка відображає, в який сезон клієнти
роблять найбільше покупок.


```python
season = df['Season'].value_counts().reset_index()
season.columns = ['Season', 'Purchase Amount (USD)']
season_order = ['Spring','Summer','Fall','Winter']
plt.figure(figsize=(8,6))
sns.barplot(data=season, x='Season', y='Purchase Amount (USD)', order=season_order, palette='bright')
plt.title('Seasonality of purchases')
plt.xlabel('Season')
plt.ylabel('Purchase Amount (USD)')
plt.show()
```

8.Графік розподілу розмірів одягу (size):
Створіть графік розподілу розмірів одягу, щоб визначити популярність
різних розмірів.

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Size', palette='pastel', order=df['Size'].value_counts().index)
plt.title('Distribution of clothing sizes')
plt.xlabel('Cloth Size')
plt.ylabel('Purchase Amount')
plt.show()
```

9.Кількість покупок відносно типу оплати (payment_method):
Порівняйте, скільки покупок робиться за різними методами оплати за

Побудуйте графік поверхні (3D surface plot) для візуалізації залежності
між трьома числовими змінними з використанням бібліотеки Seaborn.
Налаштуйте відображення графіка, включаючи назву графіку, підписи
осей і інші необхідні параметри для кращого розуміння візуалізації.
Додайте кольорову легенду.
𝑧 = 𝑠𝑖𝑛 √𝑥
2 + 𝑦
2, 𝑥 ∈ [−5; 5], 𝑦 ∈ [−5; 5]

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5,5,100)
y = np.linspace(-5,5,100)
x,y = np.meshgrid(x,y)
z = np.sin(np.sqrt(x**2 + y**2))
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection ='3d')
sur = ax.plot_surface(x,y,z, cmap='viridis')

ax.set_title('Surface graphic')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```
